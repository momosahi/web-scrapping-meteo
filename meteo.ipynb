```python
def get_pays(url):
    request = requests.get(url)
    soup = bs(request.text, "html.parser")
    base = "https://www.historique-meteo.net"
    # extraction des pays
    pays_tags = soup.find_all("div", {"class":"item-text"})
    # récupération des urls pour chaque pays
    url_pays_tags = soup.find_all("a", {"class":"hover-wrap fancybox"}, href=True)
    url_pays = [base + tag["href"] for tag in url_pays_tags]
    # récupération du nom de chaque pays
    nom_pays = [tag.get_text(strip=True) for tag in pays_tags]
    infos = [(url, nom) for (url, nom) in zip(url_pays, nom_pays)]
    return infos
```

```python
def get_region(url):
    request = requests.get(url)
    soup = bs(request.text, "html.parser")
    base = "https://www.historique-meteo.net"
    # extraction des régions
    region_tags = soup.find("div").find_all_next("a", {"class":"list-group-item"}, href=True)
    # recupération des urls
    url_regions = [base + tag["href"] for tag in region_tags]
    # récupération des noms des régions (des villes en fait)
    nom_regions = ["".join(tag.get_attribute_list("title")).strip()[19:] for tag in region_tags]
    infos = [(url, nom) for (url, nom) in zip(url_regions, nom_regions)]
    # on ne garde que les données par villes (on exclut celles par années)
    return infos[15:]
```

```python
def split_on_first_digit(s):
    """" spliter une chaine de caractères sur le premier digit qui apparait."""
    for i, char in enumerate(s):
        if char.isdigit():
            return s[:i], s[i:]
    return s, ""

s = 'Vitesse du vent19km/h20km/h21km/h'
result = split_on_first_digit(s)
print(result)
```

```python
import re
import unicodedata

def slugify(string):
    # Normalize the string to remove accents and diacritics
    normalized_string = unicodedata.normalize('NFKD', string)
    # Replace non-alphanumeric characters with a hyphen
    slugified_string = re.sub(r'[^\w\s-]', '', normalized_string).strip().lower()
    # Replace spaces with a hyphen
    slugified_string = re.sub(r'[-\s]+', '-', slugified_string)
    return slugified_string
```

```python
def get_day_data(url):
    request = requests.get(url)
    soup = bs(request.text, "html.parser")
    # extraction des kpis
    kpis = soup.find("table").find_all("tr")[1:]
    kpis = [split_on_first_digit(kpi.get_text().strip())[0] for kpi in kpis]
    # suppression du dernier élement qui n'est pas utile pour nous ici
    kpis.pop()
    # Pour mettre dans le bon format avec slugify
    kpis = [slugify(kpi) for kpi in kpis]
    # extractions des valeurs des kpis
    values = soup.find("table").find_all("td", {"class":"text-center bg-primary"})[1:]
    values = [value.get_text() for value in values]
    return (kpis, values)
```

```python
def get_data(url, years=[]):
    # Retrieve all regions and their links for the chosen country
    regions = get_region(url)
    all_data = []

    for year in years:
        # Stop at June 30th for the year 2022
        if year == 2022:
            months_range = range(1, 7)
        else:
            months_range = range(1, 13)

        for month in months_range:
            for region_url, region_name in regions:
                # Retrieve data for each day of the month
                days_range = range(1, 32) if month in [1, 3, 5, 7, 8, 10, 12] else range(1, 31) if month != 2 else range(1, 29)

                for day in tqdm(days_range, desc=f"{year}-{month}"):
                    # Retrieve data for the specific day and region
                    kpis, values = get_day_data(f"{region_url}/{year}/{month:02}/{day:02}")
                    data = dict(zip(kpis, values))
                    data["Date"] = f'{year}/{month:02}/{day:02}'
                    # Add the data to the list
                    all_data.append(data)

    # Create a DataFrame from the list of dictionaries
    df = pd.DataFrame(all_data)

    return df
```

```python
url = "https://www.historique-meteo.net/afrique/cote-d-ivoire"
years = [2022]
df = get_data(url, years=years)
```
